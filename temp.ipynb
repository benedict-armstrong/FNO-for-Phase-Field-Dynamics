```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
IC_TYPE = "fourier"
EPSILON = 0.02
```

```python
data = np.load(f"../../data/Foundational Models for Phase-Field Dynamics/test_allen_cahn_{IC_TYPE}_{EPSILON}.npy")
```

```python
data.shape
```

```python
fig, axs = plt.subplots(5, 2, figsize=(10, 20))

for i, (ax, s) in enumerate(zip(axs.flatten(), data)):
    for t_i, t in enumerate(s):
        x = np.linspace(-1, 1, len(t))
        ax.plot(x, t, label=f"t={t_i}")

    # print(t)

plt.legend()
plt.show()
```
