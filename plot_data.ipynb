```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as mplstyle
mplstyle.use('fast')
```

```python
ic_types = ['piecewise', 'gmm', 'fourier']
```

```python
var = len(ic_types)

fig, axs = plt.subplots(var, 6, figsize=(20, 10))


for i, row in enumerate(axs):
    ic = ic_types[i]

    data = np.load(f"data/train_allen_cahn_{ic}.npy")
    sample_ids = np.random.choice(len(data), 6, replace=False)

    for j, ax in enumerate(row):
        sample = data[sample_ids[j]]
        epsilon = sample[0, 0, -1]

        ax.set_title(f"ic={ic}, epsilon={epsilon}")

        # print(sample.shape)

        for t, x_t in enumerate(sample[..., 0]):
            dt = sample[t, 0, 1]
            x = np.linspace(-1, 1, len(x_t))
            ax.plot(x, x_t, label=f"t={dt:.2e}")
        ax.legend()

plt.tight_layout()
plt.show()
```
